```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://v:localhost@localhost:5432/h1n1')
```

```python
train = pd.read_csv("Data/training_set_features.csv")
test= pd.read_csv("Data/test_set_features.csv")
traininglabels = pd.read_csv("Data/training_set_labels.csv")
```

```python
train.to_sql('train', engine, index=False)
```

```python
test.to_sql('test', engine, index=False)
```

```python
traininglabels.to_sql('traininglabels', engine, index=False)
```

```python
query = 'SELECT * FROM train;'
train = pd.read_sql(query, engine)

train.head(2)
```

```python
query = 'SELECT * FROM traininglabels;'
traininglabels = pd.read_sql(query, engine)

traininglabels.head(2)
```

```python
query = 'SELECT * FROM train FULL OUTER JOIN traininglabels ON train."respondent_id" = traininglabels."respondent_id";'
df=pd.read_sql(query, engine)
df.head(2)
```

```python
query = 'SELECT * FROM train WHERE "doctor_recc_h1n1" IS NOT NULL AND "doctor_recc_seasonal" IS NOT NULL AND "health_insurance" IS NOT NULL;'
df_select=pd.read_sql(query, engine)
```
